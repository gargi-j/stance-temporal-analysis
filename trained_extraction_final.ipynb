```python
import pandas as pd
import langid
```

```python
data = pd.read_csv("D:\\Minor 1\\Dataset\\mycsv_v8.csv")
```

```python
#Sorting in ascending order
data = data.sort_values(by='OR_user_screen_name',ascending=True)
```

```python
data
```

```python
user_counts = data['OR_user_screen_name'].value_counts().reset_index()
user_counts.columns = ['OR_user_screen_name', 'count']

# Filter user IDs that occur more than x times
filtered_user_ids = user_counts[(user_counts['count'] >20) & (user_counts['count'] <50)]['OR_user_screen_name']

# Create a new DataFrame with data for those user IDs
filtered_df = data[data['OR_user_screen_name'].isin(filtered_user_ids)]
```

```python
filtered_df
```

```python
filtered_df = filtered_df.dropna(subset=['OR_user_full_text'])
```

```python
filtered_df
```

```python
#Function to detect language
def detect_language(text):
    if text is not None:
        lang, _ = langid.classify(text)
        return lang
    else:
        return 'unknown'

# Apply language detection to each row and create a new column 'Language'
filtered_df['Language'] = filtered_df['OR_user_full_text'].apply(detect_language)

# Filter rows that are either in Hindi or English
df = filtered_df[filtered_df['Language'] == 'en']


print("Filtered DataFrame:")
print(df)
```

```python
df.to_csv("tr.csv")
```
