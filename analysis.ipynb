```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

# Assuming 'sorted_df' is your DataFrame

# Convert 'OR_user_Tcreated_at' to datetime format
sorted_df['OR_user_Tcreated_at'] = pd.to_datetime(sorted_df['OR_user_Tcreated_at'])

# Iterate over each user in sorted_df and plot sorted tweets
for user, user_df in sorted_df.groupby('OR_user_screen_name'):
    # Sort the tweets for the current user by 'OR_user_Tcreated_at'
    sorted_user_df = user_df.sort_values(by='OR_user_Tcreated_at')
    
    # Plotting the changes in labels over time for the current user
    plt.figure(figsize=(10, 4))  # Adjust the figure size as needed
    plt.plot(sorted_user_df['OR_user_Tcreated_at'], sorted_user_df['label'], marker='o', linestyle='-')
    plt.xlabel('Time (OR_user_Tcreated_at)')
    plt.ylabel('Label')
    plt.title(f'Label Changes Over Time for User: {user}')
    
    # Format x-axis to show datetime values
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M:%S'))
    plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
    
    # Rotate x-axis labels
    plt.xticks(rotation=45, ha='right')
    plt.savefig(f'plot_user_{user}.png', bbox_inches='tight')
    plt.show()
```
